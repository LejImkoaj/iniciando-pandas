```python
df_group = df.groupby('Aprobado')['Nota'].mean()  # Calcula la media de 'Nota' agrupada por 'Aprobado'
df_group.name = 'Media_Nota'
df_group
```

```python
df_join = df.join(df_group, on=['Aprobado'], how='inner')  # Une el DataFrame original con el DataFrame agrupado
df_join.head()
```

```python
df_join[['Media_Nota', 'Edad']].plot()  # Grafica las columnas 'Media_Nota' y 'Edad' del DataFrame unido
```
