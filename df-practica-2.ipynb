El equipo de ML ha llegado con algunas demandas de última hora que debemos resolver en este momento de análisis exploratorio. Estas demandas son las siguientes:

    Calcular el promedio de habitaciones por departamento.

    Verificar cuántas colonias únicas existen en nuestra base de datos.

    Analizar qué colonias tienen el promedio de alquiler más alto.

    Crear un gráfico de barras horizontales que muestre las 5 colonias con los promedios de alquiler más altos.


```python
import pandas as pd
```

```python
datos = pd.read_csv('https://gist.githubusercontent.com/ahcamachod/a572cfcc2527046db93101f88011b26e/raw/ffb13f45a79d31223e645611a119397dd127ee3c/alquiler.csv', sep=';') # Carga el archivo CSV desde una URL y lo guarda en un DataFrame llamado 'datos'
datos
```

```python
datos.head()
```

```python
datos['Habitaciones'].mean()
```

```python
datos.Tipo.unique()
```

```python
datos.groupby('Tipo')['Habitaciones'].mean().sort_values() # Agrupa los datos por tipo de inmueble, calcula el promedio de habitaciones para cada tipo y los ordena de menor a mayor
```

```python
habitaciones_departamentos = datos.query('Tipo == "Departamento"')[['Habitaciones']].mean() # Filtra los datos para obtener solo los departamentos y calcula el promedio de habitaciones
habitaciones_departamentos
```

    Verificar cuántas colonias únicas existen en nuestra base de datos.

 

```python
datos.head()
```

```python
datos.Colonia.unique()
```

```python
contar_colonias_unicas = datos[['Colonia']].nunique()  # Cuenta el número de colonias únicas en la columna 'Colonia'
contar_colonias_unicas
```

```python
len(datos['Colonia'].unique()) # Cuenta el número de colonias únicas en la columna 'Colonia' utilizando la función len() y unique()
```

    Analizar qué colonias tienen el promedio de alquiler más alto.



```python
datos.head()
```

```python
datos.Colonia.unique()
```

```python
promedio_colonias = datos.groupby('Colonia')[['Valor']].mean().sort_values(by='Valor', ascending=False).head(5)  # Agrupa los datos por colonia, calcula el promedio de alquiler, ordena de mayor a menor y selecciona las 5 primeras
promedio_colonias
```

    Crear un gráfico de barras horizontales que muestre las 5 colonias con los promedios de alquiler más altos.

```python
promedio_colonias.plot(kind='barh', figsize=(10,5), color='green', title='Colonias con los promedios de alquiler más altos')  # Crea un gráfico de barras horizontales con personalización adicional
```
