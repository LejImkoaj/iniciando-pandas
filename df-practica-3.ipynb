```python
import pandas as pd
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/ahcamachod/807a2c1cf6c19108b2b701ea1791ab45/raw/fb84f8b2d8917a89de26679eccdbc8f9c1d2e933/alumnos.csv')
print(df)
```

1- Verifica si la base de datos contiene datos nulos y, en caso de tenerlos, realiza el tratamiento de estos datos nulos de la manera que consideres más coherente con la situación.

```python
df.isnull()
print(df)
```

```python
df.isnull().sum()
df = df.fillna(0)
```

2- Los estudiantes "Alicia" y "Carlos" ya no forman parte del grupo. Por lo tanto, elimínalos de la base de datos.

```python
eliminar_estudiantes = df.query('Nombre == "Alicia" | Nombre == "Carlos"').index #
df.drop(eliminar_estudiantes, axis=0, inplace=False) # eliminar filas de Alicia y Carlos
```

3- Aplica un filtro que seleccione solo a los estudiantes que fueron aprobados.

```python
df.head()
```

```python
# Paso 1: Estandarizar la columna 'Aprobado' (convertir "Verdadero" a True y "Falso" a False)
df['Aprobado'] = df['Aprobado'].replace({'Verdadero': True, 'False': False}).astype(bool)
```

```python
alumnos_aprobados = df[df['Aprobado'] == True] # filtro para alumnos aprobados
print(alumnos_aprobados)
```

4- Guarda el DataFrame que contiene solo a los estudiantes aprobados en un archivo CSV llamado "alumnos_aprobados.csv".

```python
alumnos_aprobados.to_csv('alumnos_aprobados.csv', index=False)
```
