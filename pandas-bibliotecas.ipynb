# **Pandas ofrece diversas funciones para importar y exportar datos en diferentes formatos. Las principales funciones de Pandas para importar datos son las siguientes:**

**read_csv():** Esta función se utiliza para leer archivos CSV (Comma Separated Values), que son archivos de texto que contienen datos separados por comas. Es posible proporcionar varios parámetros para personalizar la lectura del CSV, como el delimitador, el encabezado, el tipo de codificación, entre otros.

**read_excel():** Esta función se utiliza para leer archivos de Excel (.xls o .xlsx) y crear un DataFrame a partir de los datos.

**read_json():** Esta función se utiliza para leer archivos JSON (JavaScript Object Notation), que son archivos de texto que contienen datos en formato de objeto JavaScript.

**read_html():** Esta función se utiliza para leer tablas HTML, que son estructuras de datos organizadas en formato de tabla en una página web.

**read_sql():** Esta función se utiliza para leer datos de una base de datos relacional, como MySQL, PostgreSQL y SQL Server. Pandas es capaz de importar datos de diferentes maneras, lo que permite ajustar parámetros como la consulta, el nombre de la tabla y el tipo de datos de las columnas.

# **Demandas de ML**

###¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
datos.head()
```

```python
datos['Valor'].mean() # mean() - calcula el promedio
```

```python
datos.groupby('Tipo').mean(numeric_only=True)
```

```python
datos.groupby('Tipo')['Valor'].mean(numeric_only=True) # groupby() - agrupa
                                                       # Agrupa por tipo - de la columna valor el promedio
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor') # Agrupa por tipo - de la columna valor el promedio
                                                             # sort_values() - ordena los valores de menor a mayor  
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor') # Se asigna a una variable
df_tipo_precio.plot(kind='barh', figsize=(12,8), color='purple') # Grafica de barras horizontales
```

**Valores promedio de alquiler por tipo de inmueble**

```python
datos.Tipo.unique() # unique() - muestra los valores unicos de la columna tipo  
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala', 
'Edificio Completo','Tienda/Salón', 'Casa Comercial', 
'Terreno Estándar', 'Cochera/Estacionamiento', 
'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial',
'Hotel', 'Loteo/Condominio', 'Industria']
```

```python
datos.query('@inmuebles_comerciales in Tipo') # Filtra los inmuebles comerciales
```

```python
datos.query('@inmuebles_comerciales not in Tipo') # Filtra los inmuebles que no son comerciales
datos
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo') # guarda en varible df y filtra los inmuebles que no son comerciales    
df.head() # Visualiza las primeras 5 celdas del nuevo dataframe df
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor') # Se asigna a una variable
df_tipo_precio.plot(kind='barh', figsize=(12,8), color='green')
```

**Porcentaje de cada tipo de inmueble presente en la base de datos**

```python
df.Tipo.unique()
```

```python
df.Tipo.value_counts() # value_counts() - cuenta la cantidad de veces que aparece cada valor unico
```

```python
df.Tipo.value_counts(normalize=True) * 100 # Porcentaje de cada tipo de inmueble presente en la base de datos   
```

```python
porcentaje_tipo = df.Tipo.value_counts(normalize=True).to_frame() # Convierte a un dataframe
porcentaje_tipo.plot(kind='bar', figsize=(12,8), color='orange', xlabel='Tipos', ylabel='Porcentaje')
```
