```python
import pandas as pd
import numpy as np
```

```python
data = {"Nombre": ["Daniel", "Isabel", "Diana", "Salome", None, "Manuela", "Sofia", "Geronimo", "Gabriela", "Nelcy"],
         "Edad": [40, 39, 39, None, 5, 6, 5, 7, 13, 63],
         "Ciudad": ["Panama", "Tulua", "Buga", "Andalucia", "Sevilla", "Trujillo", "Riofrio", "Bolivar", "San pedro", np.nan ]
         }
df = pd.DataFrame(data)
df
```

# Limpieza de datos

### Rellenar los valores faltantes

```python
df_fill = data.fillna(
    {
        "Edad": data["Edad"].mean(),
        "Nombre": "Desconocido"  
    } 
)
df_fill
```
