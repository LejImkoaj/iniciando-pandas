```python
data = {"Nombre": ["Daniel", "Isabel", "Diana", "Salome", None, "Salome", "Diana", "Daniel", "Salome", "Gaby"],
         "Edad": [40, 39, 39, None, 5, 6, 39, 40, 13, 63],
         "Ciudad": ["Panama", "Tulua", "Buga", "Andalucia", "Sevilla", "Andalucia", "Buga", "Panama", "Andalucia", np.nan ]
         }
df = pd.DataFrame(data)
df
```

# Limpieza de datos

### Rellenar los valores faltantes

```python
rellenar_datos = df.fillna({
    'Nombre': 'Desconocido',
    'Edad': df['Edad'].mean().round(),
    'Ciudad': 'Desconocida'
})
rellenar_datos
```

### Otra manera de rellenar los valores Vacios:
### Interpolar Valores

```python
df_interpolar = df.copy()
df_interpolar["Edad"] = df["Edad"].interpolate()
df_interpolar
```

### Eliminar Filas con valores Faltantes

```python
df_sin_nan = df.dropna()
df_sin_nan
```

### Reemplazar valores Especificos de alguna Columna

```python
def_reemplazar = df.replace(
    {
        "Ciudad" : {None: "Desconocida"}
    }
)

def_reemplazar
```

### Eliminar Duplicados => Cuando las Filas son Identicas en todas las columnas

```python
df_sin_duplicados = df_sin_nan.drop_duplicates()
df_sin_duplicados   
```

### Para Eliminar Duplicados por el nombre de la Columna

```python
df_sin_duplicados = df_sin_nan.drop_duplicates(subset="Nombre", keep="first").reset_index(drop=True)
df_sin_duplicados
```

### Renombrar las Columnas

```python
df_renombrado = df_sin_duplicados.rename(columns={"Nombre": "Name", "Edad": "Age", "Ciudad": "City"})
df_renombrado
```

### Ordenar Columnas

```python
columnas_ordenadas = ["City", "Name", "Age"]
df_ordenado = df_renombrado[columnas_ordenadas]
df_ordenado
```

### Transformacion de Datos

```python
def cuadrado(x): # Función para calcular el cuadrado de un número
    return x ** 2
df_renombrado["Edad_Cuadrado"] = df_renombrado["Age"].apply(cuadrado).round() # Aplicar la función a la columna "Age"
df_renombrado
```
